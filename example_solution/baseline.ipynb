```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df_train = pd.read_csv('train.csv')
```

```python
cols_wo_nans = []
for col in df_train.columns:
    if (df_train[col].isnull()).sum() == 0:
        cols_wo_nans.append(col)
```

```python
print(', '.join(cols_wo_nans))
```

```python
df_train[cols_wo_nans].head()
```

```python
X_train = df_train[['Var57', 'Var73', 'Var113']]
y_train = df_train['Target']
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
df_to_predict = pd.read_csv('data_predict.csv')
X_pred = df_to_predict[['Var57', 'Var73', 'Var113']]
y_pred = clf.predict_proba(X_pred)[:,1]
```

```python
submission = pd.DataFrame({'ID': df_to_predict['ID'].to_list(), 'Target': y_pred})
```

```python
df_to_predict['ID']
```

```python
submission.to_csv('sample_submission.csv', index=False)
```
